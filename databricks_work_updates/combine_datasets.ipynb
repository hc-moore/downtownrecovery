```python
in_both = np.intersect1d(us_city_index_with_features.columns.values, updated_can_vars.columns.values)
in_both
```

```python

# change them back to original variables so modeling is easier
updated_us_vars = pd.concat([us_city_index_with_features[in_both]]).drop(columns = ["Unnamed: 0", "Unnamed: 0.1"])

updated_us_vars.head()
```

```python
var_in_both2_us = np.intersect1d(updated_us_vars.drop(columns = 'city').columns.values, original_model_data.drop(columns = 'city').columns.values)
in_both2_us= np.intersect1d(updated_us_vars.columns.values, original_model_data.columns.values)
```

```python
both_data = updated_us_vars[in_both2_us].melt(id_vars = 'city', value_name = 'updated').merge(original_model_data[in_both2_us].melt(id_vars = 'city', value_name = 'original'), how = 'inner', on = ['city', 'variable'])
```

```python
both_data['difference'] = both_data['updated'] - both_data['original']
both_data['pct_of_og'] = 100 * abs(both_data['difference'] / both_data['original'])
changed_variables = both_data[both_data['variable'].str.contains('_city') & ~both_data['variable'].str.contains('density') & ~both_data['variable'].str.contains('pct_hisp')]
changed_variables.groupby('variable')[['difference', 'pct_of_og']].mean().sort_values(by = 'pct_of_og', ascending = False)
```

```python
var_in_both2_can = np.intersect1d(updated_can_vars_subset.drop(columns = 'city').columns.values, original_model_data.drop(columns = 'city').columns.values)
in_both2_can= np.intersect1d(updated_can_vars_subset.columns.values, original_model_data.columns.values)
```

```python
both_data_can = updated_can_vars_subset[in_both2_can].melt(id_vars = 'city', value_name = 'updated').merge(original_model_data[in_both2_can].melt(id_vars = 'city', value_name = 'original'), how = 'inner', on = ['city', 'variable'])
```

```python
both_data_can['difference'] = both_data_can['updated'] - both_data_can['original']
both_data_can['pct_of_og'] = 100 * abs(both_data_can['difference'] / both_data_can['original'])
both_data_can.sort_values(by = 'difference', ascending = False)
```

```python

# now join this to original model data

original_model_data_subset_us = original_model_data.drop(columns = var_in_both2_us)


original_model_data_subset_us.columns
```
