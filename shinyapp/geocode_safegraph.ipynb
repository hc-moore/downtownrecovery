```python
import pandas as pd
```

```python
your_data = pd.read_csv("~/data/downtownrecovery/your_data_mar_15_2022_0846am.csv")
                             
```

```python
toronto_data = your_data[your_data['postal_code'].str[0] == 'M'].reset_index().drop(columns = 'index').reset_index()
toronto_data['index'] = toronto_data['index'] + 1
toronto_data.head()
```

```python
toronto_addresses = toronto_data[['index', 'street_address', 'city', 'region', 'postal_code', 'iso_country_code']].drop_duplicates().rename(columns = {'index' : 'OBJECTID', 'street_address' : 'Address', 'city': 'City', 'region': 'Region', 'postal_code': 'Postal', 'iso_country_code':'CountryCode'})
```

```python
to_geocode = toronto_addresses.apply(lambda x : x.to_dict(), axis = 1).values.tolist()
to_geocode
```

```python
from arcgis.gis import GIS
from arcgis.geocoding import get_geocoders, batch_geocode
```

```python
gis = GIS("https://cal.maps.arcgis.com", client_id = "PIoVkRmnSdgfuhON")
```

```python
# use the first of GIS's configured geocoders
geocoder = get_geocoders(gis)[0]
```

```python
print("For current geocoder:")
print(" - MaxBatchSize: " + str(geocoder.properties.locatorProperties.MaxBatchSize))
print(" - SuggestedBatchSize: " + str(geocoder.properties.locatorProperties.SuggestedBatchSize))
```

```python
results2 = batch_geocode(to_geocode)
```

```python
results2
```

```python
geocoded_addresses = pd.DataFrame.from_dict(results2)
geocoded_addresses
```

```python
# x is longitude, y is latitude
geocoded_addresses_fixed = geocoded_addresses.join(pd.json_normalize(geocoded_addresses['attributes'])).drop('attributes', axis = 'columns')
geocoded_addresses_fixed
```

```python
pd.set_option('display.max_column', None)
geocoded_addresses_fixed.head()
```

```python
# join back to toronto_data on index <-> ResultID
toronto_data_geocoded = toronto_data.merge(geocoded_addresses_fixed, how = 'left', left_on = 'index', right_on = 'ResultID')
toronto_data_geocoded.head()
```
